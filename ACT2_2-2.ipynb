```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss

# Fetch historical data for INTC
ticker = 'INTC'
start_date = '2019-11-29'
end_date = '2025-02-25'
data = yf.download(ticker, start=start_date, end=end_date)

# Calculate moving averages
data['MA9'] = data['Close'].rolling(window=9).mean()
data['MA30'] = data['Close'].rolling(window=30).mean()

# Perform ADF test
adf_result = adfuller(data['Close'].dropna())
adf_output = pd.Series(adf_result[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
for key, value in adf_result[4].items():
    adf_output['Critical Value (%s)' % key] = value

# Perform KPSS test
kpss_result, kpss_p_value, _, critical_values = kpss(data['Close'].dropna(), regression='c', nlags="auto")
kpss_output = pd.Series([kpss_result, kpss_p_value], index=['Test Statistic', 'p-value'])
for key, value in critical_values.items():
    kpss_output['Critical Value (%s)' % key] = value

# Plotting the data
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['MA9'], label='9-Day MA')
plt.plot(data['MA30'], label='30-Day MA')
plt.title('INTC Stock Price with 9-Day and 30-Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Save the plot
plt.savefig('INTC_Technical_Analysis.png')

# Print the test results
print("ADF Test Results:")
print(adf_output)
print("\nKPSS Test Results:")
print(kpss_output)
```
